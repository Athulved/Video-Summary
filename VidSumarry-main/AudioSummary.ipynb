```python
import subprocess

def extract_audio(video_path, audio_output_path):
    command = f"ffmpeg -i {video_path} -q:a 0 -map a {audio_output_path} -y"
    subprocess.call(command, shell=True)
    print(f"Audio extracted and saved to {audio_output_path}")
extract_audio("/Users/mymac/Projects/VidSumarry/testVideo2.mp4","/Users/mymac/Projects/VidSumarry/aud.mp3")
```

```python
import whisper

def transcribe_audio(audio_path):
    # Load the Whisper model
    model = whisper.load_model("base")  # You can use "small", "medium", or "large" for better accuracy
    
    # Transcribe the audio
    result = model.transcribe(audio_path)
    
    # Return the transcribed text
    return result["text"]
audio_path="/Users/mymac/Projects/VidSumarry/aud.mp3"
r=transcribe_audio(audio_path)
print(r)
```

```python
from transformers import pipeline

def summarize_text(text, max_length=150):
    # Load the summarization pipeline
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    
    # Generate the summary
    summary = summarizer(text, max_length=max_length, min_length=30, do_sample=False)
    
    # Return the summarized text
    return summary[0]["summary_text"]
output=summarize_text(r)
print(output)
```
